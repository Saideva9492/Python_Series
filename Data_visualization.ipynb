# Univariate Analysis

### Univariate Continous data- Histogram

```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
plt.hist(x='mpg',data=mt_cars_df)
plt.ylabel("Frequency")
plt.xlabel('MPG')
plt.title("MPG Distribution",size=15)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(x='disp',data=mt_cars_df)
plt.ylabel("Frequency")
plt.xlabel('disp')
plt.title("Displacement Distribution",size=15)
plt.show()
```

### Univariate Discrete data - Countplot


```python
sns.countplot(x="cyl",data=mt_cars_df,palette="inferno",)
```

```python
plt.pie(x=mt_cars_df["cyl"].value_counts(),autopct="%1.0f%%",labels=[8,4,6],explode=[0.02,0,0.1])
```

```python
plt.boxplot(x="mpg",data=mt_cars_df)
plt.show()
```

### Bivariate Analysis continous vs continous - scatter plot

```python
mt_cars_df.head()
```

```python
sns.scatterplot(x='mpg',y='wt',data= mt_cars_df)
plt.show()
```

### Bivariarte (Discrete vs Continous) Barplot

```python
plt.figure(figsize=(10,8))
sns.barplot(x="hp",y='wt',data=mt_cars_df,width=0.5,)
plt.xlabel("Horse Power")
plt.ylabel("Weight")
plt.title('Horse Power Vs Weight')
mt_cars_df.head()
```

### Bivariarte (Discrete vs Discrete) Barplot

```python
pd.crosstab(index=mt_cars_df['cyl'],columns=mt_cars_df['gear'])
```

### Multivariate
